# **Handling Mouse Event by Using OpenCV**

```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow

# Initialize a blank image
img = np.zeros((512, 512, 3), np.uint8)

# Mouse callback function
def draw(event, x, y, flags, param):
    if event == cv2.EVENT_LBUTTONDBLCLK:  # Left double click
        cv2.circle(img, (x, y), 100, (125, 0, 255), 5)

    if event == cv2.EVENT_RBUTTONDBLCLK:  # Right double click
        cv2.rectangle(img, (x, y), (x + 100, y + 75), (125, 125, 255), 2)

# Create a named window and set the mouse callback
cv2.namedWindow("res")
img=np.zeros((512,512,3),np.uint8)
cv2.setMouseCallback("res", draw)

while True:
    # Display the image
    cv2_imshow(img)

    # Wait for the 'q' key to exit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Clean up and close the window
cv2.destroyAllWindows()
```

# **Create a fucntion which help to find cordinate of any pixel and its color without using mouse**

```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow

# Load the image
img = cv2.imread('/content/Thor.jpeg')

if img is None:
    print("Error: Image not found. Please check the path.")
else:
    cv2_imshow(img)  # Display the image initially
    print("Click on the image to get coordinates or color values.")
    print("Type 'exit' to stop the program.")

    while True:
        # Ask the user for coordinates
        user_input = input("Enter coordinates in the format 'x,y' or type 'exit': ")

        if user_input.lower() == 'exit':
            break

        try:
            # Parse the coordinates
            x, y = map(int, user_input.split(','))

            # Ensure coordinates are within bounds
            if 0 <= x < img.shape[1] and 0 <= y < img.shape[0]:
                # Get the color values at the specified coordinates
                b, g, r = img[y, x]
                color_bgr = f"Color at ({x}, {y}): B={b}, G={g}, R={r}"
                print(color_bgr)

                # Draw coordinates on the image
                cv2.putText(img, f"{x}, {y}", (x, y), cv2.FONT_HERSHEY_PLAIN, 1, (155, 125, 100), 2)
                cv2_imshow(img)  # Refresh the image display
            else:
                print("Coordinates are out of bounds. Please try again.")
        except ValueError:
            print("Invalid input format. Please enter in the format 'x,y'.")

# Clean up: No need to destroy windows in Colab, as the kernel handles it.
print("Program exited.")
```
