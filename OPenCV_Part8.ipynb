#Thresholding

**Thresholding in OpenCV is a technique that converts an image to binary form, by setting pixel values to either the maximum or minimum, based on a threshold value.**

# Advantages of Thresholding in OpenCV:

**Simplifies Image Analysis:** Converts complex images to simple binary, reducing data complexity.

**Highlights Key Features:** Makes objects or regions of interest more distinct.

**Speeds Up Processing:** Reduces computational load for further image processing
tasks.

**Enhances Object Detection:**Useful for segmenting objects from the background.

**Improves Accuracy:** Particularly in conditions with consistent lighting, it enhances pattern recognition.

# **3 types of Threholding**

# **Simple Thresholding:** Applies a fixed threshold globally to all pixels, setting each pixel to max or min based on a specified threshold.

THRESH_BINARY, THRESH_BINARY_INV, THRESH_TRUNC, THRESH_TOZERO, THRESH_TOZERO_INV – apply a fixed threshold globally.



```python
#Threholding is a segmentation techique which is use to separate selected object from an image.

#Image Thresholding -  If pixel value is greater than a threshold value
#it is assigned one value (may be white),
#else it is assigned another value (may be black).
#thresholding is use to subtract image from background
#Thresholding is of  3 type -  Simple thresholding, Adaptive thresholding, Otsu’s thresholding
#image should be in gray scale
#simple thresholding(img,pixel_thresh,_max_thresh_pixel,style)
#it return 2 values - one is random data , second is threshold


# Import necessary libraries
import cv2
from google.colab.patches import cv2_imshow  # For displaying images in Colab
import numpy as np

# Load the image in grayscale
img = cv2.imread("/content/btw.jpeg", 0)  # Ensure image path is correct
img = cv2.resize(img, (300, 300))

# Display the original image
cv2_imshow(img)

# Apply different thresholding techniques
_, th1 = cv2.threshold(img, 50, 255, cv2.THRESH_BINARY)
_, th2 = cv2.threshold(img, 50, 255, cv2.THRESH_BINARY_INV)
_, th3 = cv2.threshold(img, 127, 255, cv2.THRESH_TRUNC)
_, th4 = cv2.threshold(img, 127, 255, cv2.THRESH_TOZERO)
_, th5 = cv2.threshold(img, 127, 255, cv2.THRESH_TOZERO_INV)

# Display the thresholded images
print("1 - THRESH_BINARY")
cv2_imshow(th1)

print("2 - THRESH_BINARY_INV")
cv2_imshow(th2)

print("3 - THRESH_TRUNC")
cv2_imshow(th3)

print("4 - THRESH_TOZERO")
cv2_imshow(th4)

print("5 - THRESH_TOZERO_INV")
cv2_imshow(th5)
```

# **Adaptive Thresholding**: Uses local neighborhood values to set different thresholds for different regions, handling images with uneven lighting.

# **ADAPTIVE_THRESH_MEAN_C** – threshold is the local mean minus a constant.

# **ADAPTIVE_THRESH_GAUSSIAN_C** – threshold is a Gaussian-weighted sum of the neighborhood minus a constant.

# **cv2.adaptiveThreshold(source, maxValue, adaptiveMethod, thresholdType, blockSize, C)**

```python
#Adaptive thresholding -
#We use it beacuse simple thresholding not able to handle
#different type of low luminous pixels
# this, the algorithm calculate the threshold for a small regions of the image.
#so we get multiple threshold for diff. regions in same image.

#Adaptive Method - It decides how thresholding value is calculated.
#cv2.ADAPTIVE_THRESH_MEAN_C : threshold value is the mean of neighbourhood area.
#cv2.ADAPTIVE_THRESH_GAUSSIAN_C : threshold value is the weighted sum of neighbourhood values where weights are a gaussian window.


#threshold (img,pixel_thresh,_max_thresh_pixel,method,style,no._of_pixel,contact_mean)
# Import necessary libraries
import cv2
from google.colab.patches import cv2_imshow  # For displaying images in Colab
import numpy as np

# Load the image in grayscale
img = cv2.imread('/content/btw1.jpeg', 0)  # Make sure the path is correct
img = cv2.resize(img, (400, 400))

# Apply Simple Thresholding
_, th1 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)  # Simple thresholding

# Apply Adaptive Thresholding - Mean
th2 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 2)

# Apply Adaptive Thresholding - Gaussian
th3 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)

# Display the original and thresholded images
print("Original Image")
cv2_imshow(img)

print("Simple Threshold - THRESH_BINARY")
cv2_imshow(th1)

print("Adaptive Threshold - ADAPTIVE_THRESH_MEAN_C")
cv2_imshow(th2)

print("Adaptive Threshold - ADAPTIVE_THRESH_GAUSSIAN_C")
cv2_imshow(th3)
```

# **Otsu’s Thresholding:** Automatically finds an optimal global threshold by minimizing intra-class variance, ideal for images with a bimodal histogram.

# **cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)**

```python
import cv2
from google.colab.patches import cv2_imshow  # For displaying images in Colab

# Load the grayscale image
img = cv2.imread('/content/btw2.jpg',0)  # Ensure the path is correct

# Apply Otsu's thresholding
# Note: threshold value is set to 0, and we specify THRESH_OTSU to let OpenCV determine the best threshold
_, otsu_thresh = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# Display the original and Otsu thresholded image
print("Original Image")
cv2_imshow(img)

print("Otsu's Thresholding")
cv2_imshow(otsu_thresh)
```
