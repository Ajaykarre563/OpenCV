#Read and display image

```python
import cv2
from google.colab.patches import cv2_imshow

# Load the image
image_path = "/content/aws.jpeg"
image = cv2.imread(image_path)

# Display the image
cv2_imshow(image)
```

# Resize the **Image**

```python
#Resize the Image
resized_image = cv2.resize(image, (400,400))
cv2_imshow(resized_image)
```

# GrayScale Image Convert

```python
#GrayScale Image Convert
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
```

# Blur Image Convert

```python
#Blur Image Convert
blur_image = cv2.GaussianBlur(image, (5,5), 0)
cv2_imshow(blur_image)
```

## Save image

```python
#Save image
cv2.imwrite("gray_image.jpg", gray_image)
```

```python

import cv2
from google.colab.patches import cv2_imshow

cap = cv2.VideoCapture("/content/AK.mp4")

# Loop through each frame in the video
while True:
  ret,frame = cap.read()
  #resize the video frame
  frame=cv2.resize(frame,(400,400))
  frame=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    # Display the frame using cv2_imshow in Colab
  cv2_imshow(frame)
  k=cv2.waitKey(5)
  # You can terminate the video frame by using "q"
  if k==ord("q") & 0xFF:
    break
cap.release()
cv2.destroyAllWindows()
```

# Capture the video frames Web Cam and save in memory.




```python
import cv2
from google.colab.patches import cv2_imshow

cap = cv2.VideoCapture(0)
#DIVX, XVID, MJPG, X264, WMV1, WMV2
fourcc = cv2.VideoWriter_fourcc(*"XVID")  # *"XVID"
#It contain 4 parameter , name, codec,fps,resolution
#output.avi drive path
output = cv2.VideoWriter("output.avi",fourcc,20.0,(640,480),0)

# Loop through each frame in the video
while cap.isOpened():
  ret,frame =cap.read()
  if ret==True:
    #frame=cv2.resize(frame,(400,400))
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    # reverse the frame
    frame=cv2.flip(frame,0)
    cv2.imshow('frame',frame)
    cv2.imshow('gray',gray)
    #fourcc,20.0,(640,480),0) last paramter indicate gray scale

    output.write(gray)
    if cv2.waitKey(1) & 0xFF == ord('q'): # press to exit
      break

cap.release()
output.release()
cv2.destroyAllWindows()
```
