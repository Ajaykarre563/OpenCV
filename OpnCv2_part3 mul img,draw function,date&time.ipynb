# **Capture Multiple Images and Store in a folder**

```python
import cv2
```

```python
import cv2
from google.colab.patches import cv2_imshow

# Load the video file
vidcap = cv2.VideoCapture("/content/Ak.mp4")

count = 0

while True:
    ret, image = vidcap.read()  # Read a frame from the video
    if ret:  # Check if the frame was successfully read
        cv2.imwrite(f"Frame{count}.jpg", image)  # Save frame as JPEG file
        cv2_imshow(image)  # Display the frame using cv2_imshow
        count += 1

        # Wait for a short period and check if 'q' is pressed to quit
        if cv2.waitKey(1) & 0xFF == ord("q"):
            break
    else:
        break  # Exit the loop if no more frames

# Release the video capture object
vidcap.release()
cv2.destroyAllWindows()
```

## **Drawing function on Images**

```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow


img = cv2.imread("/content/Thor.jpeg")
img = cv2.resize(img,(600,700))


# Creating a blank image
#img = np.zeros([512, 512, 3], np.uint8) + 255  # For a white background
#img = np.zeros([512, 512, 3], np.uint8)  # Creates a black background

# Drawing a line
#Here line accept 5 parameter (img,starting,ending,color,thickness)
img = cv2.line(img, (0, 0), (200, 200), (154, 92, 255), 8)  # Fixed color value

# Drawing an arrowed line
#arrowed line accept also accpet 5 parameter  (img,starting,ending,color,thickness)
img = cv2.arrowedLine(img, (0, 125), (255, 255), (255, 0, 125), 10)

# Drawing a rectangle
#Rectangle - accept parameter(img,start_co,end_co,colot ,thickness)
img = cv2.rectangle(img, (384, 10), (510, 128), (128, 0, 255), 8)

# Drawing a filled circle
#circle - accept(img,star_co,radius,color,thickness)
img = cv2.circle(img, (447, 125), 63, (214, 255, 0), -1)  # Changed thickness to -1 for filling

# Adding text
#puttext -accept(img,text,start_co,font,fontsize,color,thickness,linetype)
font = cv2.FONT_ITALIC
img = cv2.putText(img, 'THOR', (20, 500), font, 4, (0, 125, 255), 10, cv2.LINE_AA)

# Drawing an ellipse
#ellipse-accept(img,start_cor,(length,height),color,thickness)
img = cv2.ellipse(img, (400, 600), (100, 50), 0, 0, 180, (155, 0, 0), 5)  # Fixed color value

# Drawing a polygon
pts = np.array([[100, 150], [200, 30], [170, 20], [50, 10]], np.int32)
pts = pts.reshape((-1, 1, 2))
img = cv2.polylines(img, [pts], True, (0, 255, 155))

# Displaying the image
cv2_imshow(img)
```

# **set data and time and figure on video**

```python
import cv2
from google.colab.patches import cv2_imshow
import datetime

# Load the video
cap = cv2.VideoCapture("/content/Ak.mp4")  # Adjust the path as needed

# Check video properties
print("Original Width:", cap.get(cv2.CAP_PROP_FRAME_WIDTH))
print("Original Height:", cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

# Set new width and height
cap.set(3, 600)  # Width
cap.set(4, 800)  # Height
print("Set Width:", cap.get(3))
print("Set Height:", cap.get(4))

while cap.isOpened():
    ret, frame = cap.read()
    if ret:
        # Prepare text and date
        font = cv2.FONT_HERSHEY_COMPLEX_SMALL
        text = f'Height: {cap.get(4)} Width: {cap.get(3)}'
        date_data = f"Date: {datetime.datetime.now()}"

        # Put text on frame
        #accept parameters(img,text,start_co,font,font_size,colot,thickness,lineType)
        frame = cv2.putText(frame, text, (10, 20), font, 1, (0, 155, 255), 1, cv2.LINE_AA)
        #accept parameters(img,date_date,start_co,font,font_size,color,thickness,lineType)
        frame = cv2.putText(frame, date_data, (20, 50), font, 1, (100, 255, 255), 1, cv2.LINE_AA)

        # Draw rectangle
        #Rectangle - accept parameter(img,start_co,end_co,colot ,thickness)
        cv2.rectangle(frame, (384, 10), (510, 128), (128, 0, 255), 8)

        # Display the frame
        cv2_imshow(frame)

        # Wait for 1 ms to check for 'q' key press
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    else:
        break

# Release resources
cap.release()
cv2.destroyAllWindows()
```
