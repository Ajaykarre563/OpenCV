**Contours in computer vision** are curves that connect points of similar intensity along object boundaries, useful for shape analysis, object detection, and recognition.

**Key Steps for Using Contours**

**Binary Image:** Convert the image to grayscale and use thresholding or edge detection (like Canny) to define edges clearly.
    
**Contour Detection:** Use findContours on a binary image where objects are white on a black background. Copy the image first, as this function modifies it.


# We have to find and draw contours as per the requirement.

```python
#Contours -
#Contours can be explained simply as a curve joining all the continuous points
#(along the boundary), having same color or intensity.

#The contours are a useful tool for shape analysis and object detection and recognition

#For better accuracy, use binary images and also apply edge detection before
#finding countours.

#findCountour function manipulate original imge so copy it before proceeding.
#findContour is like finding white object from black background.
#so you must turn image in white and background is black.


import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
img = cv2.imread("/content/logo.png")

# Convert to grayscale
img1 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply thresholding
ret, thresh = cv2.threshold(img1, 150, 255, 0)

# Find contours
cnts, hier = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

# Print contour information
print("Contours:", cnts)
print("Total contours:", len(cnts))
print("Hierarchy:\n", hier)

# Draw contours on the original image
img_contour = cv2.drawContours(img.copy(), cnts,-1, (176, 10, 15), 4)

# Function to display images inline
def display_image(image, title="Image"):
    # Convert BGR to RGB for displaying with matplotlib
    if len(image.shape) == 3:
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(image, cmap='gray' if len(image.shape) == 2 else None)
    plt.title(title)
    plt.axis('off')
    plt.show()

# Display images
display_image(img, "Original Image")
display_image(img1, "Grayscale Image")
display_image(thresh, "Threshold Image")
display_image(img_contour, "Contours Drawn")
```

# **Moments in OpenCV**

**To calculate the moments of an image or contour in OpenCV, you can use the cv2.moments() function. This function returns a dictionary containing spatial moments, central moments, and central normalized moments. Here's a quick breakdown:**

1.**Spatial Moments (m00, m10, m01, etc.):** These describe the distribution of pixels within a region.
    
2.**Central Moments:** These are invariant to translation, meaning they describe the shape regardless of its position in the image.
    
3.**Contours -** These are invariant to scale, making them ideal for comparing shapes of different sizes.

```python
#Contours can be explained simply as a curve joining all the continuous points
#(along the boundary), having same color or intensity.

#The contours are a useful tool for shape analysis and object detection and recognition

#For better accuracy, use binary images and also apply edge detection before
#findig countours.

#findCountour function manipulate original imge so copy it before proceeding.
#findContour is like finding white object from black background.
#so you must turn image in white and background is black.


# Import necessary libraries
import cv2
import numpy as np
from google.colab.patches import cv2_imshow  # Use cv2_imshow for image display in Colab

# Specify the path to your image in Google Drive
image_path = '/content/sha.jpg'  # Replace with your image's path in Google Drive

# Load and process the image
img = cv2.imread(image_path)
if img is None:
    print("Error: Image not found. Please check the path.")
else:
    img = cv2.resize(img, (600, 700))

    # Convert to grayscale
    img1 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Apply binary thresholding
    ret, thresh = cv2.threshold(img1,238, 255, cv2.THRESH_BINARY_INV)

    # Find contours
    #findcontour(img,contour_retrival_mode,method)
    cnts, hier = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    #Here cnts is a list of contours. ANd each contour is an array with x, y cordinate
    #hier variable called hierarchy and it contain image information.
    print("Number of contours:", len(cnts))
    print("Hierarchy:\n", hier)

    # Loop over each contour
    for c in cnts:
        # Calculate moments for each contour
        M = cv2.moments(c)
        print("Moments:", M)

        # Calculate center if area (m00) is not zero
        if M["m00"] != 0:
            cX = int(M["m10"] / M["m00"])
            cY = int(M["m01"] / M["m00"])

            # Draw contour and mark the center
            cv2.drawContours(img, [c], -1, (0, 255, 0), 2)
            cv2.circle(img, (cX, cY), 7, (255, 255, 255), -1)
            cv2.putText(img, "center", (cX - 20, cY - 20), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2)
        else:
            print("Contour with zero area detected, skipping center calculation.")

    # Display images in Colab
    print("Original Image with Contours and Centers:")
    cv2_imshow(img)
    print("Grayscale Image:")
    cv2_imshow(img1)
    print("Thresholded Image:")
    cv2_imshow(thresh)
```

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow  # Use cv2_imshow for image display in Colab

# Provide the image path manually or use input() to prompt for it
image_path = input("Enter the full path of the image (e.g., /content/your_image.jpg): ")

# Load the image from the provided path
img = cv2.imread(image_path)

# Check if the image was loaded successfully
if img is None:
    print("Error: Image not found. Please check the path.")
else:
    # Resize and convert to grayscale
    img = cv2.resize(img, (600, 700))
    img1 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Apply binary thresholding
    ret, thresh = cv2.threshold(img1, 238, 255, cv2.THRESH_BINARY_INV)

    # Find contours
    cnts, hier = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    print("Number of contours:", len(cnts))

    area1 = []
    # Loop over each contour
    for c in cnts:
        # Calculate moments for each contour
        M = cv2.moments(c)

        # Calculate center if area (m00) is not zero
        if M["m00"] != 0:
            cX = int(M["m10"] / M["m00"])
            cY = int(M["m01"] / M["m00"])

            # Calculate area of the contour
            area = cv2.contourArea(c)
            area1.append(area)

            if area < 10000:
                # Contour Approximation to reduce the number of points
                epsilon = 0.1 * cv2.arcLength(c, True)
                approx = cv2.approxPolyDP(c, epsilon, True)

                # Convex Hull of the contour
                hull = cv2.convexHull(approx)

                # Draw bounding rectangle around the convex hull
                x, y, w, h = cv2.boundingRect(hull)
                img = cv2.rectangle(img, (x, y), (x + w, y + h), (125, 10, 20), 5)

            # Draw contour and mark the center
            cv2.drawContours(img, [c], -1, (50, 100, 50), 2)
            cv2.circle(img, (cX, cY), 7, (255, 255, 255), -1)
            cv2.putText(img, "center", (cX - 20, cY - 20), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2)

    # Display images using cv2_imshow in Colab
    cv2_imshow(img)  # Display the original image with contours
    cv2_imshow(img1)  # Display the grayscale image
    cv2_imshow(thresh)  # Display the thresholded image

    # Wait for any key press to close the images
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```
